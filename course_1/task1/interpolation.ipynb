```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

$ g(x) = c_0 + c_1 x $

```python
A = np.array([[1, 1], [1, 3]])
b = np.array([f(1), f(15)])
c = linalg.solve(A, b)
print(c)
```

```python
def g(x):
    return c[0] + c[1] * x
```

```python
x = np.arange(0, 16, 0.1)
f_v = np.vectorize(f)
g_v = np.vectorize(g)
plt.plot(x, f_v(x))
plt.plot(x, g_v(x))
plt.show()
```

```python
A = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b = np.array([f(1), f(8), f(15)])
c = linalg.solve(A, b)
print(c)
```

```python
def g(x):
    return c[0] + c[1] * x + c[2] * x**2
```

```python
x = np.arange(0, 16, 0.1)
g_v = np.vectorize(g)
plt.plot(x, f_v(x))
plt.plot(x, g_v(x))
plt.show()
```

$$ g(x) = c_0 + c_1 x + c_2 x^2 + c_3 x^3 = \begin{pmatrix} c_0 & c_1 & c_2 & c_3 \end{pmatrix} \begin{pmatrix} 1 \\ x \\ x^2 \\ x^3 \end{pmatrix} $$

```python
def h(k, deg=4):
    return np.array([k**i for i in range(deg)])
```

```python
points = [1, 4, 10, 15]

A = np.array([h(p) for p in points])
b = np.array([f(p) for p in points])
c = linalg.solve(A, b)
print(c)
```

```python
def g(x):
    return c.dot([1, x, x**2, x**3])
g_v = np.vectorize(g)
plt.plot(x, f_v(x))
plt.plot(x, g_v(x))
plt.show()
```
